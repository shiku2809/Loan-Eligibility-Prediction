 Outlier Detection using Z-Score
</p>

<!-- Outlier Treatment Steps -->
<ul style="line-height:1.8; color:#34495E; font-family:Verdana, sans-serif; font-size:105%;">
  <li> Calculate Z-scores for numerical features such as  <b>LoanAmount</b></li>
  <li> Mark values as outliers if their absolute Z-score exceeds a threshold (typically 3.0)</li>
  <li> Remove or cap the identified outliers to reduce skewness and improve model generalization</li>
</ul>

<!-- Summary -->
<p style="font-family:Verdana, sans-serif; font-size:105%; color:#2C3E50;">
Using Z-score for outlier detection ensures that extreme anomalies are addressed without arbitrarily trimming valid data, leading to a cleaner and more reliable dataset for modeling.
</p>


```python
len_1 = len(df)

from scipy.stats import zscore
df['z_score'] = zscore(df['LoanAmount'])
df = df[df['z_score'].abs() <=3]  # keep rows within 3 standard deviations
df.drop(columns='z_score', inplace=True)

len_2 = len(df)
print("Percentage of data removed in outlier detection:", round(((len_1 - len_2) * 100) / len_1, 2), "%")
```

</p>


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, roc_curve, roc_auc_score
)
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline   # only for CV


def classify(model, X, y):
    # 1. Train-test split with stratification
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, stratify=y, test_size=0.25, random_state=42
    )

    # 2. Apply SMOTE on training data only
    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

    # 3. Scale the train and test sets
    scaler = MinMaxScaler()
    X_train_resampled = scaler.fit_transform(X_train_resampled)
    X_test = scaler.transform(X_test)

    # 4. Train the model
    model.fit(X_train_resampled, y_train_resampled)

    # 5. Evaluate on test set
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1] 

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob) if y_prob is not None else None
    cm = confusion_matrix(y_test, y_pred)

    print(f"üîπ Accuracy:  {acc:.4f}")
    print(f"üîπ Precision: {prec:.4f}")
    print(f"üîπ Recall:    {rec:.4f}")
    print(f"üîπ F1 Score:  {f1:.4f}")
    if auc is not None:
        print(f"üîπ ROC‚ÄìAUC:   {auc:.4f}")

    # 6. Cross-validation with same process (SMOTE + Scaling + Model)
    pipeline = Pipeline([
        ('smote', SMOTE(random_state=42)),
        ('scaler', MinMaxScaler()),
        ('model', model)
    ])
    scores = cross_val_score(pipeline, X, y, cv=5, scoring='accuracy')
    print("üî∏ Cross-validation Accuracy:", np.mean(scores) * 100)

    # 7. Confusion Matrix
    print("\nüîπ Confusion Matrix:\n", cm)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    # 8. ROC Curve
    if y_prob is not None:
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        plt.figure(figsize=(6, 5))
        plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {auc:.4f})")
        plt.plot([0, 1], [0, 1], 'k--', label="Random Guess")
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.title("ROC Curve")
        plt.legend(loc="lower right")
        plt.show()

    return model
```
